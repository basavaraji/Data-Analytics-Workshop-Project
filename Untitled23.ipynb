```python
import matplotlib.pyplot as plt

department = data.groupby("Department")["Percentage"].mean()
Colors = ["blue","black","orange","red","green"]

plt.figure(figsize=(8,5))
plt.bar(department.index, department.values,color = Colors, alpha = 0.4)

plt.title("Department wise Average Percentage")
plt.xlabel("Department")
plt.ylabel("Average Percentage")
plt.grid()
plt.show()
```
